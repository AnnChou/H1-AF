```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
```

```python
train_pixel_sz_HC_file = csv_path + 'training_set_pixel_size_and_HC.csv'
train_pixel_sz_HC_data = pd.read_csv(train_pixel_sz_HC_file)
train_pixel_sz_HC_data.head()
```

```python
# Create X 
features = ['pixel size(mm)']
X = train_pixel_sz_HC_data[features]
X.head()
```

```python
#Set target
target_name ='head circumference (mm)'
y = train_pixel_sz_HC_data[target_name]
```

```python
# Split into validation and training data
#Train Random forest model on train_x train_y
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1)

# Define a random forest model
rf_model = RandomForestRegressor(random_state=1)
rf_model.fit(train_X, train_y)
rf_val_predictions = rf_model.predict(val_X)
rf_val_mae = mean_absolute_error(rf_val_predictions, val_y)

print("Validation MAE for Random Forest Model: {:,.0f}".format(rf_val_mae))
```

```python
# To improve accuracy, create a new Random Forest model which I will train on all training data
rf_model_on_full_data = RandomForestRegressor(random_state=1)

# fit rf_model_on_full_data on all data from the training data
rf_model_on_full_data.fit(X, y)
```

Now, read the file of "test" data, and apply my model to make predictions.

```python
# path to file I will use for predictions
test_pixel_sz_file = csv_path + 'test_set_pixel_size.csv'
# read test data file using pandas
test_pixel_sz_data = pd.read_csv(test_pixel_sz_file)
test_pixel_sz_data.head()
```

```python
# create test_X which comes from test_data but includes only the columns you used for prediction.
# The list of columns is stored in a variable called features
test_X = test_pixel_sz_data[features]

# make predictions which we will submit. 
test_preds = rf_model_on_full_data.predict(test_X)
print(test_X)
```

```python
# --- utils  ----
# purpose:  a FUNCTION that exports a tensor object as a file from a Jupyter Notebook:

import numpy as np
import pickle
import tensorflow as tf

def export_tensor_as_file(tensor, file_path):
    if isinstance(tensor, np.ndarray):
        np.save(file_path, tensor)
    elif isinstance(tensor, tf.Tensor):
        tf.io.write_file(file_path, tf.io.serialize_tensor(tensor))
    else:
        with open(file_path, 'wb') as file:
            pickle.dump(tensor, file)

            
# # Assuming my tensor object is named 'tensor'
export_tensor_as_file(tensor, 'tensor_file.npy')


```
